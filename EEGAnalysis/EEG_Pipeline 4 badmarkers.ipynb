```python
#Cool function for entering all subelements
def traverse(o, tree_types=(list, tuple)):
    if isinstance(o, tree_types):
        for value in o:
            for subvalue in traverse(value, tree_types):
                yield subvalue
    else:
        yield o
        
#fixing markers
def fix_markers(xdf_path, info=False, channels=None):
    
    streams = xdf.load_xdf(xdf_path)
    
    for i in range(len(streams[0])):
        if "Exp_group_data" in streams[0][i]['info']['name'][0]:
            fixed_unitystr_time_series = streams[0][i]['time_series']
            
            newmarkers = []
            for t2 in fixed_unitystr_time_series:
                for e in t2:
                    if "alter_pitch" in e:
                        newmarkers.append(0)
                    if "SphereCollision" in e:
                        newmarkers.append(1)
                    if "Missed" in e:
                        newmarkers.append(2)
                    if "BombCollision" in e:
                        newmarkers.append(3)
                    if "Sphere_wave" in e:
                        newmarkers.append(4)
                    if "Camera_swap" in e:
                        newmarkers.append(5)
            
            fixed_unitystr_time_series = newmarkers
            return fixed_unitystr_time_series

fix_markers("C:\\Recordings\\CurrentStudy\\exp1\\DimiTrainDay1.xdf")
```

```python


def get_raw_and_events(xdf_path, info=False, channels=None):
    
    print("-------- Fetching data... --------------")
    streams = xdf.load_xdf(xdf_path)       
    unityEvs = False
    
    for i in range(len(streams[0])):
        if "EEG" in streams[0][i]['info']['name'][0]:
            eeg_time_series     = streams[0][i]['time_series']
            eeg_time_stamps     = streams[0][i]['time_stamps']
            eeg_first_timestamp = np.float(streams[0][i]['footer']['info']['first_timestamp'][0])
            
        if "Exp_group_data" in streams[0][i]['info']['name'][0]:
            fixed_unitystr_time_series     = streams[0][i]['time_series']
            
            #cleaning up markers
            
            newmarkers = []
            for t2 in fixed_unitystr_time_series:
                for e in t2:
                    if "alter_pitch" in e:
                        newmarkers.append(0)
                    if "SphereCollision" in e:
                        newmarkers.append(1)
                    if "Missed" in e:
                        newmarkers.append(2)
                    if "BombCollision" in e:
                        newmarkers.append(3)
                    if "Sphere_wave" in e:
                        newmarkers.append(4)
                    if "Camera_swap" in e:
                        newmarkers.append(5)
            
            fixed_unitystr_time_series = newmarkers
            
            
            
            unitystr_time_stamps     = streams[0][i]['time_stamps']
            unitystr_first_timestamp = np.float(unitystr_time_stamps[0])
            fixed_unitystr_time_series= np.squeeze(fixed_unitystr_time_series).astype(int)
            unityEvs = True
            
            print("checking first time stamps, Unity:", unitystr_first_timestamp, )

            
    # Sampling frequency of the recorded eeg data
    sfreq = np.float(streams[0][1]['info']['nominal_srate'][0]) # here it is 500, so leave it
    
    # 32 channels:
    ch_names = ('P7', 'P4', 'Cz', 'Pz', 'P3', 'P8', 'O1', 'O2', 'T8', 'F8', 'C4', 'F4', 'Fp2', 'Fz', 'C3', 'F3', 'Fp1', 'T7', 'F7', 'Oz', 'PO4', 'FC6', 'FC2', 'AF4', 'CP6', 'CP2', 'CP1', 'CP5', 'FC1', 'FC5', 'AF3', 'PO3')
    # Create the info file
    info = mne.create_info(ch_names, sfreq, ch_types='eeg' )
        
    if info is True:        
        print(info)
    # Convert first time to sample - why do we need it?
    first_samp = int(eeg_first_timestamp*sfreq)
    
    print("-------- Creating raw... --------------")
    # current code
    raw = mne.io.RawArray(eeg_time_series.T, info)
    
    # montage
    montage = mne.channels.read_montage('standard_1020')
    raw.set_montage(montage)
    
    # Create a ne stimulation channels
    info         = mne.create_info(['STI'], raw.info['sfreq'], ch_types=['stim'])
    # Fill it with an vector of zeros
    stim_data    = np.zeros((1, len(raw.times)))
    stim_raw     = mne.io.RawArray(stim_data, info) # do we leave this first_samp?
    # Add the new channel to your dataset
    raw.add_channels([stim_raw], force_update_info=False)
    
    print("-------- Adding events... --------------")
    
    firstcol_events = (unitystr_time_stamps*1).astype('int') # sfreq different here, may alter
    secondcol_events = np.zeros((len(fixed_unitystr_time_series)))
    thirdcol_events = np.squeeze(fixed_unitystr_time_series)
    
    if unityEvs is True:

        # You need to build an array of events
        events = np.array([firstcol_events, secondcol_events, thirdcol_events]).astype('int')
        
        raw.load_data()
        # Add the  events 
        raw.add_events(events.T, stim_channel='STI')
        ### Re-reference your data
        raw.set_eeg_reference()    
        raw.notch_filter(np.arange(60, 181, 60), fir_design='firwin')    
        events = mne.find_events(raw, stim_channel='STI', output='onset', shortest_event=1,
                             min_duration=1/raw.info['sfreq'])
        print('Number of events found: %d' % len(events))

    else:
        raw.set_eeg_reference()    
        raw.notch_filter(np.arange(60, 181, 60), fir_design='firwin')    
        events = None
    
    print("-------- Raw and events done! --------------")
    return (raw, events)

get_raw_and_events("C:\\Recordings\\CurrentStudy\\exp1\\DimiTrainDay1.xdf", info=False, channels=None)
```
